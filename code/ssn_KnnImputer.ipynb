```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import KNNImputer
imp = KNNImputer(n_neighbors=5, weights="uniform") # or ‘distance’
```

```python
data.replace('?', np.nan, inplace=True)
datanp = data.to_numpy()
datanp[5]
```

```python
dataimp = imp.fit_transform(datanp)
dataimp[5]
```

```python
dataimp[:,:4] = np.round(dataimp[:,:4])
dataimp[:,5] = np.round(dataimp[:,5])
dataimp[:,7:-3] = np.round(dataimp[:,7:-3])
dataimp[5]
```

```python
dfnew = pd.DataFrame(dataimp, columns=data.columns)
dfnew.to_csv("knn_low_ssn.csv", index=False)
#dfnew.to_csv("knn_high_ssn.csv", index=False)
```
